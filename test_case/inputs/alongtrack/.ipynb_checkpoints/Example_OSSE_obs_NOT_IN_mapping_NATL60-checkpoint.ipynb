## CASE STUDY 2: INDEPENDENT ALONG-TRACK

__INTRODUCTION__

We here propose to show the behaviour of the spectral coherence approach in a case study based on NATL60 simulation.
The input comes from hourly SSH data from a 1 year long period of the run NATL60-CJM165.
We used the SWOTsimulator to sample alongtrack data (nadir altimeter) on this simulation. The constellation is similar to the 20030101-20031231 period: Jason 1, Envisat, Geosat2, Topex-Tandem. SWOTsimulator gives us the true SSH (SSH model) and a pseudo-obs SSH (SSH model with white noise). These pseudo-obs SSH are then used in the DUACS mapping.

__CASE STUDY #2:__
- the along-track of reference (for comparison) is __Geosat2__. In this case study #2, this along-track __is NOT included__ in the mapping.
- the SSH map produced are generated from 3 nadir altimeters TPN, J1, EN with the DUACS system
- we focus on the region in the north-atlantic bassin (10째x10째 bax centered near 300째E-30째N)

PYTHON MODULE

```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, interpolate
from matplotlib.ticker import FormatStrFormatter
```

__READ THE SSH SEGMENTS__

_Note_: The SSH segments were generated using same algorithm as used in the manuscript. There are 800km long, referenced by their mean latitude and mean longitude and they overlap over 25%

__NOTATION__

- __true_SSH_model__: alongtrack SSH from model
- __noised_SSH_model__: alongtrack SSH from model including white noise
- __mapped_SSH__: SSH mapped with the DUACS algorithm using 3 altimeters data, and interpolated onto the alongtrack SSH from model path
- __intrumental_error__: noised_SSH_model minus true_SSH_model
- __mapping_error__: mapped_SSH minus true_SSH_model

```python
nc = Dataset('/Users/mballaro/Public/scuba_grid_cls/segment_independent_alongtrack_ssh_obs_g2.nc','r')
noised_SSH_model = nc.variables['sla_segment'][:]
mapped_SSH = nc.variables['sla_study_segment'][:]
dx = nc.variables['resolution'][:]
nc.close()
nperseg = 135
```

```python
nc = Dataset('/Users/mballaro/Public/scuba_grid_cls/segment_independent_alongtrack_ssh_model_g2.nc','r')
true_SSH_model = nc.variables['sla_segment'][:]
nc.close()
```

```python
intrumental_error = noised_SSH_model - true_SSH_model
mapping_error = true_SSH_model - mapped_SSH
```

```python
plt.figure(figsize=(15, 6))
plt.xlabel("Time")
plt.ylabel("Signal Amplitude")
plt.plot(true_SSH_model, c='b', label='ssh_model')
plt.plot(noised_SSH_model, c='lightblue', label='ssh_obs')
plt.plot(mapped_SSH, c='orange', label='ssh_map')
plt.ylim(-1, 1)
plt.legend(loc='best')
plt.grid()
```

__SPECTRAL CONTENT OF EACH SIGNAL__

```python
freq, psd_true_SSH_model = signal.welch(true_SSH_model, fs=1/dx, nperseg=nperseg, scaling='density', noverlap=0)
_, psd_noised_SSH_model = signal.welch(noised_SSH_model, fs=1/dx, nperseg=nperseg, scaling='density', noverlap=0)
_, psd_mapped_SSH = signal.welch(mapped_SSH, fs=1/dx, nperseg=nperseg, scaling='density', noverlap=0)
_, psd_intrumental_error = signal.welch(intrumental_error, fs=1/dx, nperseg=nperseg, scaling='density', noverlap=0)
_, psd_mapping_error = signal.welch(mapping_error, fs=1/dx, nperseg=nperseg, scaling='density', noverlap=0)
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
plt.plot(1/freq, psd_true_SSH_model, c='b', label='true_SSH_model')
plt.plot(1/freq, psd_noised_SSH_model, c='lightblue', label='noised_SSH_model')
plt.plot(1/freq, psd_mapped_SSH, c='orange', label='mapped_SSH')
plt.plot(1./freq, psd_intrumental_error, c='r', label='intrumental_error')
plt.plot(1./freq, psd_mapping_error, c='magenta', label='mapping_error')
plt.gca().invert_xaxis()
plt.yscale('log')
plt.xscale('log')
ax.xaxis.set_major_formatter(FormatStrFormatter('%.f'))
ax.xaxis.set_minor_formatter(FormatStrFormatter('%.f'))
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('PSD (m2/km)', fontweight='bold')
plt.legend(loc='best')
plt.grid(which='both')
```

The figure illustrate a goog agreement of the SSH along-trcak and SSH map power spectrum to about 300km. Below 300 km the SSH maps are less energetic linked to the filtering properties of te DUACS mapping system. The figure also shows that the instrumental noise is larger than the map error for lenghscales < . Above, the mapping error dominates. The successive bumps in the PSD of the SSH_mapped signal below 50 km are linked to the linear interpolation of the SSH maps onto the along-track location. Hence, in the evaluation it is not necessary to consider wavelenght < 50 km, and we are thus in the case where SSH_mapped error are larger than the intrumental error.

__SPECTRAL RATIO__

```python
fig, ax = plt.subplots(figsize=(15, 3))
plt.plot(1./freq, psd_mapped_SSH/psd_noised_SSH_model, c='b', label='psd_mapped_SSH/psd_noised_SSH_model')
#plt.ylim(0, 2.5)
plt.xscale('log')
plt.gca().invert_xaxis()
ax.xaxis.set_major_formatter(FormatStrFormatter('%.f'))
ax.xaxis.set_minor_formatter(FormatStrFormatter('%.f'))
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('Spectral Ratio', fontweight='bold')
plt.hlines(0.5, xmin=0, xmax=500, linestyle='--')
plt.legend()
plt.grid(which='both')
```

__CROSS SPECTRUM__

We compute the cross spectrum between mapped_SSH and the noise_SSH_model. This is used later in the transfer function

```python
freq, cross_psd_mapped_SSH_noised_SSH_model = signal.csd(mapped_SSH, noised_SSH_model, fs=1./dx, nperseg=nperseg, noverlap=0)
```

__SPECTRAL COHERENCE__

```python
# Compute coherence
freq, coh_true_SSH_model_intrumental_error = signal.coherence(true_SSH_model, intrumental_error, 
                                                                  fs=1./dx, nperseg=nperseg, noverlap=0)
_, coh_noised_SSH_model_intrumental_error = signal.coherence(noised_SSH_model, intrumental_error, 
                                                                 fs=1/dx, nperseg=nperseg, noverlap=0)
_, coh_mapped_SSH_noised_SSH = signal.coherence(mapped_SSH, noised_SSH_model, fs=1/dx, nperseg=nperseg, noverlap=0)
_, coh_mapped_SSH_true_SSH_model = signal.coherence(mapped_SSH, true_SSH_model, fs=1/dx, nperseg=nperseg, noverlap=0)
_, coh_mapped_SSH_intrumental_error = signal.coherence(mapped_SSH, intrumental_error, fs=1/dx, 
                                                       nperseg=nperseg, noverlap=0)
_, coh_mapped_SSH_mapping_error = signal.coherence(mapped_SSH, mapping_error, fs=1/dx, 
                                                       nperseg=nperseg, noverlap=0)
```

```python
fig, ax = plt.subplots(figsize=(15, 3))
plt.plot(1./freq, psd_true_SSH_model/psd_intrumental_error, c='b', label='PSD_ssh_model / PSD_noise')
plt.plot(1./freq, psd_noised_SSH_model/psd_intrumental_error, c='g', label='PSD_ssh_obs / PSD_noise')
plt.plot(1./freq, psd_true_SSH_model/psd_mapping_error, c='orange', label='PSD_ssh_model / PSD_mapping_error')
plt.plot(1./freq, psd_mapped_SSH/(psd_mapping_error+psd_intrumental_error), c='r', 
         label='PSD_map / (PSD_mapping_error + PSD_noise)')
plt.plot(1./freq, psd_mapped_SSH/psd_mapping_error, c='grey', label='PSD_map / PSD_mapping_error')
plt.plot(1./freq, psd_mapped_SSH/psd_intrumental_error, c='magenta', label='PSD_map / (PSD_noise)')
plt.ylim(0, 2.5)
plt.xscale('log')
plt.gca().invert_xaxis()
ax.xaxis.set_major_formatter(FormatStrFormatter('%.f'))
ax.xaxis.set_minor_formatter(FormatStrFormatter('%.f'))
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('Signal-Noise-Ratio (SNR)', fontweight='bold')
plt.legend()
plt.grid(which='both')

fig, ax = plt.subplots(figsize=(15, 3))
plt.plot(1./freq, cross_psd_mapped_SSH_noised_SSH_model/psd_noised_SSH_model, c='b', label='')
#plt.ylim(0, 2.5)
plt.xscale('log')
plt.gca().invert_xaxis()
ax.xaxis.set_major_formatter(FormatStrFormatter('%.f'))
ax.xaxis.set_minor_formatter(FormatStrFormatter('%.f'))
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('Transfer function', fontweight='bold')
plt.hlines(0.5, xmin=0, xmax=500, linestyle='--')
plt.legend()
plt.grid(which='both')

#plt.subplot(212)
fig, ax = plt.subplots(figsize=(15, 3))
plt.plot(1./freq, coh_true_SSH_model_intrumental_error, c='b', label='C ssh_model/noise')
plt.plot(1./freq, coh_noised_SSH_model_intrumental_error, c='g', label='C ssh_obs/noise')
plt.plot(1./freq, coh_mapped_SSH_intrumental_error, c='r', label='C ssh_map/noise')
plt.plot(1./freq, coh_mapped_SSH_noised_SSH, c='orange', label='C ssh_map/ssh_obs', lw=6)
plt.plot(1./freq, coh_mapped_SSH_true_SSH_model, c='magenta', label='C ssh_map/ssh_model')
plt.hlines(0.5, xmin=0, xmax=500, linestyle='--')
#plt.plot(1./freq, coh_signal_unnoised_signal, c='lightblue', label='signal1 filtered')
plt.gca().invert_xaxis()
plt.grid(which='both')
plt.xscale('log')
plt.legend()
ax.xaxis.set_major_formatter(FormatStrFormatter('%.f'))
ax.xaxis.set_minor_formatter(FormatStrFormatter('%.f'))
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('Magnitude Squared Coherence', fontweight='bold')

fig, ax = plt.subplots(figsize=(15, 3))
plt.plot(1./freq, psd_mapping_error/psd_noised_SSH_model, c='b', label='psd_map_error/psd_noised_SSH_model')
plt.plot(1./freq, psd_mapping_error/psd_mapped_SSH, c='r', label='psd_map_error/psd_mapped_SSH')
plt.hlines(0.5, xmin=0, xmax=500, linestyle='--')
plt.ylim(0,2)
plt.gca().invert_xaxis()
plt.grid(which='both')
plt.legend()
plt.xscale('log')
ax.xaxis.set_major_formatter(FormatStrFormatter('%.f'))
ax.xaxis.set_minor_formatter(FormatStrFormatter('%.f'))
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('Ratio PSD_Error / PSD_alongtrack', fontweight='bold')
```

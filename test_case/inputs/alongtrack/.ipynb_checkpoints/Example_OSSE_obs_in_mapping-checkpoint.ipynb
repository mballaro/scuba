```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, interpolate
```

```python
dx = 7.   # km
max_spectrum_lenght = 500  # km
nperseg = int(max_spectrum_lenght/dx)
initial_date = 19358.
final_date = 19722.
```

```python
nc = Dataset('/Users/mballaro/en/BOOST-SWOT_EN_nadir_c02.nc','r')
#nc = Dataset('/Users/mballaro/j1/BOOST-SWOT_J1_nadir_c02.nc','r')
ssh_model = nc.variables['ssh_model'][750:1300]
ssh_obs = nc.variables['ssh_obs'][750:1300]
lon = nc.variables['lon'][750:1300]
lat = nc.variables['lat'][750:1300]
time = nc.variables['time'][750:1300]
nc.close()

# time in days
time = initial_date + time/(24.*3600.)
print(time[0], time[-1])
noise = ssh_obs - ssh_model
```

```python
nc = Dataset('/Users/mballaro/map_sla_obs_NATL60_en_j1_tpn.nc','r')
ssh_map = nc.variables['Grid_0001'][:, :, :].swapaxes(1,2)/100.
lon_map = nc.variables['NbLongitudes'][:]
lat_map = nc.variables['NbLatitudes'][:]
nc.close()
time_map = np.arange(initial_date, final_date + 1, 1)
```

```python
time_index = np.where(time_map == int(time[0]))[0]
ssh_map = ssh_map[time_index, : , :]
finterp_map2alongtrack = interpolate.RegularGridInterpolator(
    [lat_map, lon_map], ssh_map[0, :, :], bounds_error=False, fill_value=None)

ssh_interpolated = finterp_map2alongtrack(np.transpose([lat, lon]))
map_error = ssh_interpolated - ssh_model
```

```python
plt.figure(figsize=(15, 6))
plt.xlabel("Time")
plt.ylabel("Signal Amplitude")
plt.plot(ssh_model, c='b', label='ssh_model')
plt.plot(ssh_obs, c='lightblue', label='ssh_obs')
plt.plot(ssh_interpolated, c='orange', label='ssh_map')
plt.legend(loc='best')
plt.grid()
```

```python
freq, psd_ssh_model = signal.welch(ssh_model, fs=1/dx, nperseg=nperseg)
_, psd_ssh_obs = signal.welch(ssh_obs, fs=1/dx, nperseg=nperseg)
_, psd_noise = signal.welch(noise, fs=1/dx, nperseg=nperseg)
_, psd_ssh_map = signal.welch(ssh_interpolated, fs=1/dx, nperseg=nperseg)
_, psd_map_error = signal.welch(map_error, fs=1/dx, nperseg=nperseg)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(1./freq, psd_ssh_model, c='b', label='ssh_model')
plt.plot(1./freq, psd_ssh_obs, c='lightblue', label='ssh_obs')
plt.plot(1./freq, psd_noise, c='r', label='noise')
plt.plot(1./freq, psd_ssh_map, c='orange', label='ssh_map')
plt.plot(1./freq, psd_map_error, c='magenta', label='ssh_map_error')
plt.gca().invert_xaxis()
plt.yscale('log')
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('PSD (m2/km)', fontweight='bold')
plt.legend(loc='best')
plt.grid()
```

```python
# Compute coherence
freq, coh_ssh_model_noise = signal.coherence(ssh_model, noise, fs=1./dx, nperseg=nperseg)
_, coh_ssh_obs_noise = signal.coherence(ssh_obs, noise, fs=1/dx, nperseg=nperseg)
_, coh_ssh_map_ssh_obs = signal.coherence(ssh_interpolated, ssh_obs, fs=1/dx, nperseg=nperseg)
_, coh_ssh_map_ssh_model = signal.coherence(ssh_interpolated, ssh_model, fs=1/dx, nperseg=nperseg)
_, coh_ssh_map_noise = signal.coherence(ssh_interpolated, noise, fs=1/dx, nperseg=nperseg)
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(211)
plt.plot(1./freq, psd_ssh_model/psd_noise, c='b', label='PSD_ssh_model / PSD_noise')
plt.plot(1./freq, psd_ssh_obs/psd_noise, c='g', label='PSD_ssh_obs / PSD_noise')
plt.ylim(0, 2)
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('Signal-Noise-Ratio (SNR)', fontweight='bold')
plt.legend()
plt.grid()

plt.subplot(212)
plt.plot(1./freq, coh_ssh_model_noise, c='b', label='C ssh_model/noise')
plt.plot(1./freq, coh_ssh_obs_noise, c='g', label='C ssh_obs/noise')
plt.plot(1./freq, coh_ssh_map_noise, c='r', label='C ssh_map/noise')
plt.plot(1./freq, coh_ssh_map_ssh_obs, c='orange', label='C ssh_map/ssh_obs')
#plt.plot(1./freq, coh_ssh_map_ssh_obs, c='magenta', label='C ssh_map/ssh_model')
plt.hlines(0.5, xmin=0, xmax=500, linestyle='--')
#plt.plot(1./freq, coh_signal_unnoised_signal, c='lightblue', label='signal1 filtered')
#plt.gca().invert_xaxis()
plt.grid()
plt.legend()
#plot.xscale('log')
plt.xlabel('Wavelenght (km)', fontweight='bold')
plt.ylabel('Magnitude Squared Coherence', fontweight='bold')
```
